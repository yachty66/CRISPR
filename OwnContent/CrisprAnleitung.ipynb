# Einführung

Du brauchst nur diese Anleitung und keine weiteren Informationen, um das Experiment erfolgreich ausführen zu können.

Im folgenden Experiment werden wir mit CRISPR DNA von Kolibakterien so verändern, sodass die Kolibakterien in einer Umgebung überleben können, in der sie normalerweise nicht überleben würden.

Cas9 ist ein Protein welches von dem Ribosom welches sich in einer Zelle befindet produziert wird

# Phase 1: Platten herstellen

1. $5.25g$ Agar in eine $250ml$ Flasche geben und 150ml Wasser dazu geben


2. Flasche stark schütteln


3. Flasche mit **nicht** fest zugedrehtem Deckel (leicht aufsitzend) in $30$ Sekunden Abständen in einer Mikrowelle erhitzen (90°-100°). Dabei die Flasche permanent beobachten, dass es zu keinem überkochen kommt. 


4. Wenn die Mischung eine klares gelb besitzt nehme die Flasche aus der Mikrowelle und lasse sie $30$ Minuten abkühlen.


5. Solange der Inhalt der Flasche noch etwas warm ist entferne die Deckel von sieben Petri Platten und befülle sie bis zur hälfte und setze den Deckel darauf. (Der inhalt wird sich härten beim kühlen)


6. Lasse die Platten für $8$ Stunden bei Raumtemperatur stehen. Wenn du danach nicht direkt das Experiment ausführst lege die Platten bei $4$° auf der Deckelseite (drehe die Platten) in den Kühlschrank.

# Phase 2: Platten mit Bakterien herstellen

1. Mit Pipette $100\mu l$ von sterilem Wasser nehmen und der DH5$\alpha$ Spitze hinzugeben


2. Die gemixte Spitze für eine Minute schütteln bis keine Bakterien mehr sichtbar sind


3. Den gesamten Inhalt ($100\mu l$) der Spitze mit einer Pipette auf **eine** Agar Platte geben


4. Mit einer Impföse die auf die Agar Platte getröpfelte Bakterienmischung verteilen 


5. Warten bis die Platte getrocknet ist, Deckel drauf setzen und die Platte so für $1-2$ Tage bei Raumtemperatur lagern bis Bakterien anfangen in weißer Farbe zu wachsen (wenn danach nicht direkt das Experiment ausgeführt wird, dann kann die Platte mit Bakterien für eine Woche im Kühlschrank überleben)


6. $100\mu l$ von Bakterien Transformationsmix mit der Pipette aus einer Spitze nehmen und in eine neue Spitze geben


7. Mit Impföse Bakterien von der vorher erstellten Platte kratzen (bis Impöse gefüllt ist) und in die Bakterien Transformationsmix Spitze geben, sodass sich alle Klumpen auflösen. Der Transformationsmix sollte trüb sein, aber so, dass man noch hindurch sehen kann. (Wiederhole den Prozess so oft wie die Anzahl der Experimente die du durchführen möchtest und platziere die Spitzen bei $4°$ im Kühlschrank, wenn du das Experiment nicht direkt ausführen möchtest)


# Phase 3: Experiment

1. $10\mu l$ von der Cas9 Spitze zu der Spitze aus Phase $2$ Punkt $7$ pipettieren. Mit dem Finger gegen die Spitze schnipsen, sodass die Tropfen zum Boden gehen.


2. Repeat the same step with gRNA


3. Repeat the same step with Template DNA


4. Die Spitze aus den letzten Schritten bei $4$° für $30$ Minuten in den Kühlschrank geben.


5. Die Spitze aus dem letztem Schritt für $30$ Sekunden in $42$° warmes geben (nur die Spitze nicht den Inhalt).


6. $1.5ml$ Leitungswasser mit Raumtemperatur in eine Mikrozentrifugen Spitze geben und solang schütteln bis sich der Inhalt der Spitze mit dem Wasser aufgelöst hat.


7. $250 \mu l$ aus der Spitze von Schritt $6$ in die Spitze von Schritt $5$ geben.


8. Spitze bei $30$°C für $4$ Stunden oder bei Raumtemperatur $12$ Stunden verwahren. (Nicht höher als $37$°)


9. Eine LB/Strep/Kan/Arab Platte aus dem Kühlschrank nehmen und bei Raumtemperatur aufwärmen und alles aus der Spitze von Schritt $8$ darauf pipettieren.


10. Mit einer Impföse die auf die Agar Platte getröpfelte Bakterienmischung verteilen und erst nach $10$ Minuten wieder den Deckel darauf geben.
